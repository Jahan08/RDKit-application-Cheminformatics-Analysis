```python
import random
from rdkit import Chem

def split_sdf(input_file, train_ratio=0.8, random_seed=42):
    """Splits an SDF file into train and test sets."""

    # Read molecules from the SDF file
    suppl = Chem.SDMolSupplier(input_file)
    mols = [mol for mol in suppl if mol is not None]

    # Shuffle the molecules
    random.seed(random_seed)
    random.shuffle(mols)

    # Split into train and test sets
    split_index = int(train_ratio * len(mols))
    train_mols = mols[:split_index]
    test_mols = mols[split_index:]

    # Write the train and test SDF files
    with Chem.SDWriter("train.sdf") as train_writer:
        for mol in train_mols:
            train_writer.write(mol)

    with Chem.SDWriter("test.sdf") as test_writer:
        for mol in test_mols:
            test_writer.write(mol)

if __name__ == "__main__":
    input_file = "/content/Aqueous_Solubility_All.sdf"  # Replace with your SDF file path
    split_sdf(input_file)
```
